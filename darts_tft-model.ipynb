## darts 모델안에서 트랜스포머를 쓰는 TFT모델 가져옴

```python
import numpy as np
import pandas as pd
from tqdm import tqdm_notebook as tqdm

import matplotlib.pyplot as plt

from darts import TimeSeries, concatenate
from darts.dataprocessing.transformers import Scaler
from darts.models import TFTModel
from darts.metrics import mape
from darts.utils.statistics import check_seasonality, plot_acf
from darts.datasets import AirPassengersDataset, IceCreamHeaterDataset
from darts.utils.timeseries_generation import datetime_attribute_timeseries
from darts.utils.likelihood_models import QuantileRegression

import warnings

warnings.filterwarnings("ignore")
import logging

logging.disable(logging.CRITICAL)
```

```python
# training_cutoff = pd.Timestamp("2020051")
# train_a, val = train.split_after(training_cutoff)
```

```python
transformer = Scaler()
train_transformed = transformer.fit_transform(train)
val_transformed = transformer.transform(val)
series_transformed = transformer.transform(train)
```

```python
covariates = datetime_attribute_timeseries(train, attribute="weekday", one_hot=False)
covariates = covariates.stack(
    datetime_attribute_timeseries(train, attribute="weekday", one_hot=False)
)
```

```python
covariates = covariates.stack(
    TimeSeries.from_times_and_values(
        times=train.time_index,
        values=np.arange(len(train)),
        columns=["linear_increase"],
    )
)
```

```python
covariates = covariates.astype(np.float32)
```

```python
 training_cutoff = pd.Timestamp("20200510")
```

```python
scaler_covs = Scaler()
cov_train, cov_val = covariates.split_after(training_cutoff)
scaler_covs.fit(cov_train)
covariates_transformed = scaler_covs.transform(covariates)
```

```python
quantiles = [
    0.01,
    0.05,
    0.1,
    0.15,
    0.2,
    0.25,
    0.3,
    0.4,
    0.5,
    0.6,
    0.7,
    0.75,
    0.8,
    0.85,
    0.9,
    0.95,
    0.99,
]
input_chunk_length = 24*7
forecast_horizon = 24*7
my_model = TFTModel(
    input_chunk_length=input_chunk_length,
    output_chunk_length=forecast_horizon,
    hidden_size=64,
    lstm_layers=1,
    num_attention_heads=4,
    dropout=0.1,
    batch_size=16,
    n_epochs=5,
    add_relative_index=False,
    add_encoders=None,
    likelihood=QuantileRegression(
        quantiles=quantiles
    ),  # QuantileRegression is set per default
    # loss_fn=MSELoss(),
    random_state=42,
    torch_device_str="cuda"
)
```

### 예측값만 확인해보기 위해 1epoch만 돌려봄

```python
my_model.fit(train_transformed, future_covariates=covariates_transformed, verbose=True)
```

```python
my_model.save_model("tft.pth.tar")
```
